```python
import time
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models, transforms
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
def load_image(img_path, max_size=400, shape=None):
    image = Image.open(img_path).convert("RGB")
    if max(image.size) > max_size:
        size = max_size
    else:
        size = max(image.size)
    if shape:
        size = shape
    transform = transforms.Compose([
        transforms.Resize(size),
        transforms.ToTensor(),
        transforms.Normalize((0.485, 0.456, 0.406),
                             (0.229, 0.224, 0.225))
    ])
    image = transform(image)[:3, :, :].unsqueeze(0)
    return image.to(device)
```

```python
def im_convert(tensor):
    image = tensor.to("cpu").clone().detach()
    image = image.numpy().squeeze()
    image = image.transpose(1, 2, 0)
    image = image * np.array((0.229, 0.224, 0.225)) + np.array((0.485, 0.456, 0.406))
    image = image.clip(0, 1)
    return image
```

```python
def gram_matrix(tensor):
    _, d, h, w = tensor.size()
    tensor = tensor.view(d, h * w)
    gram = torch.mm(tensor, tensor.t())
    return gram
```

```python
def get_vgg():
    vgg = models.vgg16(pretrained=True).features.to(device).eval()
    for param in vgg.parameters():
        param.requires_grad_(False)
    return vgg
```

```python
def vgg_features(x, model):
    layers = {
        '0': 'conv1_1',
        '5': 'conv2_1',
        '10': 'conv3_1',
        '19': 'conv4_1',
        '21': 'conv4_2',  # content
        '28': 'conv5_1'
    }
    features = {}
    for name, layer in model._modules.items():
        x = layer(x)
        if name in layers:
            features[layers[name]] = x
    return features
```

```python
def get_resnet():
    resnet = models.resnet50(pretrained=True).to(device).eval()
    for param in resnet.parameters():
        param.requires_grad_(False)
    return resnet
```

```python
def resnet_features(x, model):
    features = {}
    x = model.conv1(x)
    x = model.bn1(x)
    x = model.relu(x)
    features['conv1'] = x
    x = model.maxpool(x)
    x = model.layer1(x)
    features['layer1'] = x
    x = model.layer2(x)
    features['layer2'] = x
    x = model.layer3(x)
    features['layer3'] = x
    return features
```

```python
def style_transfer(model, extractor, content_img, style_img, content_layer, style_layers, steps=500, lr=0.005):
    target = content_img.clone().requires_grad_(True).to(device)
    optimizer = optim.Adam([target], lr=lr)

    content_features = extractor(content_img, model)
    style_features = extractor(style_img, model)
    style_grams = {layer: gram_matrix(style_features[layer]) for layer in style_layers}

    content_weight = 1
    style_weight = 1e6

    content_losses = []
    style_losses = []
    total_losses = []

    for step in range(steps):
        target_features = extractor(target, model)
        content_loss = torch.mean((target_features[content_layer] - content_features[content_layer]) ** 2)

        style_loss = 0
        for layer in style_layers:
            target_feature = target_features[layer]
            target_gram = gram_matrix(target_feature)
            style_gram = style_grams[layer]
            d, h, w = target_feature.shape[1:]
            layer_loss = torch.mean((target_gram - style_gram) ** 2) / (d * h * w)
            style_loss += layer_loss

        total_loss = content_weight * content_loss + style_weight * style_loss

        optimizer.zero_grad()
        total_loss.backward()
        optimizer.step()

        content_losses.append(content_loss.item())
        style_losses.append(style_loss.item())
        total_losses.append(total_loss.item())

    return target, content_losses, style_losses, total_losses
```

```python
content_img = load_image("1.jpg")
style_img = load_image("2.jpg", shape=content_img.shape[2:])

results = {}
```

```python
vgg = get_vgg()
start = time.time()
vgg_output, vgg_c_list, vgg_s_list, vgg_t_list = style_transfer(vgg, vgg_features, content_img, style_img, "conv4_2",
                                                 ["conv1_1", "conv2_1", "conv3_1", "conv4_1", "conv5_1"])
vgg_time = time.time() - start
results['VGG16'] = {
    'output': im_convert(vgg_output),
    'content_loss': vgg_c_list[-1],
    'style_loss': vgg_s_list[-1],
    'total_loss': vgg_t_list[-1],
    'time': vgg_time,
    'content_losses': vgg_c_list,
    'style_losses': vgg_s_list,
    'total_losses': vgg_t_list
}
```

```python
resnet = get_resnet()
start = time.time()
res_output, res_c_list, res_s_list, res_t_list = style_transfer(resnet, resnet_features, content_img, style_img, "layer3",
                                                 ["conv1", "layer1", "layer2"])
res_time = time.time() - start
results['ResNet50'] = {
    'output': im_convert(res_output),
    'content_loss': res_c_list[-1],
    'style_loss': res_s_list[-1],
    'total_loss': res_t_list[-1],
    'time': res_time,
    'content_losses': res_c_list,
    'style_losses': res_s_list,
    'total_losses': res_t_list
}
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.title("Content")
plt.imshow(im_convert(content_img))
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("VGG16 Output")
plt.imshow(results['VGG16']['output'])
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("ResNet50 Output")
plt.imshow(results['ResNet50']['output'])
plt.axis("off")
plt.tight_layout()
plt.savefig("visual_comparison.jpg")
plt.show()
```
